```python
a = b = [1,2,3,]

b[2] = 6

print(f"a: {a}\nb : {b}")

c = d = "james"

d = "john"

print(f"c: {c}\nd : {d}")

e = f = ("hi", "hello", "python")

f = list(f)
f[2] = "ROS2"
f = tuple(f)

print(f"e: {e}, f: {f}")

g = h = 1

h = 2

print(f"g: {g}, h: {h}")
```

Python에서 변수의 ***메모리 참조(공유)*** 여부는 변수에 저장된 **객체의 종류(가변/불변)** 에 따라 달라집니다.
당신이 작성한 코드의 각각의 예는 매우 좋은 비교 사례이므로, 아래에서 **어떤 경우가 참조를 공유하고 어떤 경우는 그렇지 않은지** 를 정확히 설명해드릴게요.

🔍 핵심 개념 정리
| 객체 타입                 | 설명                                         | 예시                                     |
| --------------------- | ------------------------------------------ | -------------------------------------- |
| **가변 객체** (mutable)   | 값이 **변경 가능**하며, 참조를 공유하면 **값도 함께 변함**      | `list`, `dict`, `set`, 사용자 정의 클래스      |
| **불변 객체** (immutable) | 값이 **변경 불가능**, 참조를 공유해도 **새 대입 시 새 객체 생성** | `int`, `float`, `str`, `tuple`, `bool` |


✅ 1. 리스트 (가변 객체) – 참조 공유됨
~~~python
a = b = [1, 2, 3]
b[2] = 6
~~~
- 리스트는 **가변 객체**

- a와 b는 **같은 리스트 객체를 참조**

- b[2] = 6은 리스트 내부 값 변경 → a도 영향받음

~~~plaintext
a: [1, 2, 6]
b: [1, 2, 6]
~~~
✅ **메모리 참조 영향 있음**

✅ **2. 문자열 (불변 객체) – 참조 끊어짐**
~~~python
c = d = "james"
d = "john"
~~~
- 문자열은 **불변 객체**

- d = "john"은 기존 문자열 "james"를 수정하는 게 아니라, **새 문자열 객체를 d에 재할당**

- c는 여전히 "james" 참조

~~~plaintext
c: james
d: john
~~~
✅ **메모리 참조 영향 없음 (재할당하면 참조가 끊어짐)**

✅ **3. 튜플 → 리스트 → 튜플 (불변 → 가변 → 불변)**
~~~python
e = f = ("hi", "hello", "python")

f = list(f)     # 튜플을 리스트로 변환 (가변)
f[2] = "ROS2"   # 리스트 내용 변경
f = tuple(f)    # 다시 튜플로 변환
~~~
- e와 f는 처음엔 같은 튜플 참조

- 이후 f를 리스트로 바꾸고 변경한 후 **다시 튜플로 바꿈** → 새로운 객체

- e는 처음의 불변 튜플 그대로 유지

~~~plaintext
e: ('hi', 'hello', 'python')
f: ('hi', 'hello', 'ROS2')
~~~
✅ **중간에 객체를 바꿨으므로 참조 영향 없음**

✅ **4. 정수 (불변 객체) – 재할당시 참조 끊어짐**
~~~python
g = h = 1
h = 2
~~~
- 정수는 **불변 객체**

- h = 2는 **기존 객체를 수정하는 게 아니라, 새 객체 2를 h에 할당**

- g는 여전히 1 참조

~~~plaintext
g: 1, h: 2
~~~
✅ **메모리 참조 영향 없음**

🧠 요약 정리
| 변수       | 자료형                    | 가변/불변        | 참조 영향   | 설명                         |
| -------- | ---------------------- | ------------ | ------- | -------------------------- |
| `a`, `b` | `list`                 | 가변           | ✅ 영향 있음 | 동일 리스트 객체 참조, 값 수정 시 같이 변함 |
| `c`, `d` | `str`                  | 불변           | ❌ 없음    | 문자열은 재할당 시 참조가 끊어짐         |
| `e`, `f` | `tuple → list → tuple` | 불변 → 가변 → 불변 | ❌ 없음    | 중간에 새로운 객체로 대체됨            |
| `g`, `h` | `int`                  | 불변           | ❌ 없음    | 정수는 새 값 대입 시 참조가 끊어짐       |


🔎 **팁: 참조 여부 확인 방법**
~~~python
print(id(a))  # 변수 a가 참조하는 객체의 주소 출력
print(id(b))  # 같다면 동일 객체 참조 중
~~~
